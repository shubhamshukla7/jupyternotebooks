```python
import numpy as np

from nltk.corpus import stopwords
from nltk.stem.porter import *

from sklearn import svm, decomposition, metrics
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

import matplotlib.pyplot as plt


stemmer = PorterStemmer()
count_vect = CountVectorizer()
tfidf_transformer = TfidfTransformer()

nltk_stop_words = stopwords.words("english")
stop_words = set(['a','able','about','across','after','all','almost','also', \
                  'am','among','an','and','any','are','as','at','be','because','been','but', \
                  'by','can','cannot','could','dear','did','do','does','either','else','ever', \
                  'every','for','from','get','got','had','has','have','he','her','hers','him',\
                  'his','how','however','i','if','in','into','is','it','its','just','least', \
                  'let','like','likely','may','me','might','most','must','my','neither','no', \
                  'nor','not','of','off','often','on','only','or','other','our','own','rather', \
                  'said','say','says','she','should','since','so','some','than','that','the', \
                  'their','them','then','there','these','they','this','tis','to','too','twas', \
                  'us','wants','was','we','were','what','when','where','which','while','who', \
                  'whom','why','will','with','would','yet','you','your'])
SPACE = " "
FRACTION_TESTING = 0.10

# Given an nd-array shuffle rows in the array 
def shuffle_ndarray(nd_arr):
    np.random.seed(0)
    indices = np.random.permutation(len(nd_arr))
    print(indices)
    return nd_arr[indices[:]]

# Given an nd-array returns a list of indices with data of string type
def get_index_by_type(nd_arr, data_type):
    
    index_str_type = []
    index_oth_type = []
    index = 0
    for element in nd_arr[0,:]:
        if type(element) is str:
            index_str_type.append(index)
        else:
            index_oth_type.append(index)
        index += 1
    return index_str_type, index_oth_type

# Takes an nd-array and a list of indices with string data and 
# concatenates the columns. Returns a list of concatenated strings
def get_data_X_str(data_X, index_list):
    result = []
    for datum in data_X:
        current_str = ""
        for index in index_list:
            current_str += datum[index] + " "
        result = np.append(result, current_str)
    return list(result)

# Removes stop words from a list of documents
def remove_stop_words(documents, stop_words):
    result = [' '.join([word for word in text.split() if word not in stop_words]) for text in documents]
    return result

# Takes a list of documents and stems words for each document
def stem(documents, stemmer):
    return [" ".join([stemmer.stem(word) for word in sentence.split(" ")]) for sentence in documents]

def train(classifier, data_X_str_train, data_y_train):
    text_clf = Pipeline([
        ('vect', CountVectorizer()),
        ('tfidf', TfidfTransformer()),
        ('clf', classifier),
        ])
    text_clf = text_clf.fit(data_X_str_train, data_y_train)
    return text_clf

def predict(clf_pipeline, data_X_str_test, data_y_test):
    return clf_pipeline.predict(data_X_str_test)

def print_performance_metrics(data_y_test, predicted):
    print("Performance Metrics: ")
    print(metrics.classification_report(data_y_test, predicted))
    print("Confusion Matrix: ")
    print(metrics.confusion_matrix(data_y_test, predicted))

def benchmark_paramC_SVM(Cs):    
#     C = np.logspace(-10, 0, 10)
    scores_mean = list()
    scores_std = list()
    for value in Cs:
        svc = svm.SVC(kernel='linear', C=value)
        scores = cross_val_score(svc, X_train_tfidf, data_y_train, n_jobs = -1)
        scores_mean.append(scores.mean())
        scores_std.append(scores.std())
    return scores_mean, scores_std

# Given param values and the mean error and its std deviation, make a plot
def plot_cv_error_against_param_value(figure_id, param_vals, scores_mean, scores_std):
    plt.figure(figure_id, figsize=(8, 5))
    plt.clf()
    plt.semilogx(param_vals, scores_mean, 'b')
    plt.semilogx(param_vals, np.array(scores_mean) + np.array(scores_std), 'g--')
    plt.semilogx(param_vals, np.array(scores_mean) - np.array(scores_std), 'g--')
    plt.ylabel('CV score')
    plt.xlabel('Parameter C')
    plt.ylim(0, 1.1)
    plt.show()
```
